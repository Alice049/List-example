```python
import sys
class Student:
  def __init__(self):
    self.id=0
    self.name=''
    self.score=0
    self.next=None

head =Student()
head.next=None

def insert_f():
  ptr=Student()
  ptr.next=None
  ptr.id=eval(input('Student id: '))
  ptr.name=input('Student name: ')
  ptr.score=eval(input('Student score: '))
  print()

  prev=head
  current=head.next
  while current !=None and current.score >=ptr.score:
    prev=current
    current=current.next
  ptr.next=current
  prev.next=ptr

def delete_f():
  if head.next==None:
    print('No Sudenet record\n')
  else:
    del_id=eval(input('Delete student id: '))
    prev =head
    current=head.next      #找出串列的尾端節點且沒有最後一項
    while current !=None and del_id !=current.id:
      prev=current
      current=current.next      #找出串列的尾端節點
    if current !=None:
      prev.next=current.next  #把prev的next指向current的下一個節點
      current.next=None  #把current節點回收
      print('\n id: %d has heen deleted \n'%(del_id))
    else:
      print('\n id: %d is not found\n'%(del_id))

def modify_f():
  if head.next==None:
    print('No student record\n')
  else:
    modify_id=eval(input('Modify student id: '))
    prev=head
    current=head.next
    while current !=None and modify_id !=current.id:
      prev=current
      current=current.next
    if current !=None:
      print('. Student id: %s'%(current.id))
      print('  Student name: %s'%(current.name))
      print('. Student socre: %d\n'%(current.score))
      #把舊的資料刪除
      prev.next=current.next
      current.next=None
      #加入新的資料
      newscore=eval(input(' Please enter new score: '))
      ptr=Student()
      ptr.next=None
      ptr.id=current.id
      ptr.name=current.name
      ptr.score=newscore
      prev=head
      current=head.next
      while current !=None and current.score >= ptr.score:
        prev=current
        current=current.next
      ptr.next=current
      prev.next=ptr
      print('Data updated successfully!\n')
    else:
      print('\n id:%d not found!\n'%(modify_id))

def display_f():
  count=0
  if head.next==None:
    print('No student record\n')
  else:
    print('%-10s %-15s %-10s'%('ID','NAME','SCORE'))
    for i in range(32):
      print('-',end='')
    print()
    current=head.next
    while current !=None:
      print('%-10d %-15s %-3d'%(current.id,current.name,current.score))
      count=count+1
      current=current.next
    for i in range(32):
      print('-',end='')
    print()
    print('Total %d record(s) found\n'%(count))

def main():
  while True:
    print('\n***********Single list operation************')
    print('            <1> Insert             ')
    print('            <2> Delete             ')
    print('            <3> Modify             ')
    print('            <4> Display            ')
    print('            <5> Exit               ')
    print('***********************************************')


    try:
      option=int(input('.   Choice:     '))
    except valueError:
      print('Not a correct number')
      print('Try again \n')

    print()
    if option ==1:
      insert_f()
    elif option ==2:
      delete_f()
    elif option ==3:
      modify_f()
    elif option ==4:
      display_f()
    elif option ==5:
      #sys.exit(0)
      break
    else:
      print('不正確的選項')

main()
```
