# **環狀鏈結串列**
若將單向鏈結串列的最後一個節點next指標指向第一個節點時，則稱為此串列為環狀串列。
環狀串列可以從任何一個節點來追蹤節點，同樣的我們也假設環狀串列第一個節點不放資料。

```python
import sys
class Student:
  def __init__(self):
    self.llink=None
    self.id=0
    self.name=''
    self.score=0
    self.rlink=None

head =Student()
head.id=0
head.name=''
head.llink=head
head.rlink=head

def insert_f():
  global head
  ptr=Student()
  ptr.id=eval(input(' Syudent id: '))
  ptr.name=input(' Student name: ')
  ptr.score=eval(input(' Student score: '))
  print()

  prev=head #將prev指向head
  current=head.rlink
  #環狀串列current是否不等於head，且current分數是否不小於ptr分數
  while current !=head and current.score >= ptr.score:
    prev=current
    current=current.rlink
  ptr.rlink=current
  ptr.llink=prev
  prev.rlink=ptr
  current.llink=ptr

def delete_f():#刪除
  if head.rlink==head:
    print('\n  No student record !! \n')

  else:
    del_id=eval(input('\nDelete student id: '))
    prev=head
    current=head.rlink
    #找出環狀串列的尾端
    while current != head and del_id !=current.id:
      prev=current
      current=current.rlink
    if head !=current:
      prev.rlink=current.rlink
      current.rlink.llink=prev
      current=None
      print('.  %d has been deleted!!\n'%(del_id))
    else:
      print('   %d not found !!\n '%(del_id))

def modify_f():
  if head.rlink==head:
    print('\n  No student record !!\n')

  else:
    modify_id= eval(input('. Modify student id:'))

    prev=head
    current=head.rlink
    while current !=head and modify_id !=current.id:
      prev=current
      current=current.rlink

    if current !=head:
      print(' \n Student id:%d'%(current.id))
      print(' Student name: %s'%(current.name))
      print(' Student score: %d'%(current.score))


      prev.rlink=current.rlink
      current.rlink.llink=prev
      current.next=None

      newscore=eval(input('Please enter new score:'))
      ptr=Student()
      ptr.next=None
      ptr.id=current.id
      ptr.name=current.name
      ptr.score=newscore
      prev=head
      current=head.rlink
      while current !=head and current.score >=ptr.score:
        prev=current
        current=current.rlink
      ptr.rlink=current
      ptr.llink=prev
      prev.rlink=ptr
    else:
      print('\n Student %d not found! \n'%(modify_id))


def display_f():
  global head
  count=0
  if head.rlink ==head:
    print('\n No student record !!\n')
  else:
    print('\n %-10s %-15s % -10s'%('ID','NAME','SCORE'))
    print('********************************************')
    current=head.rlink
    while current !=head:
      print('%-10d %-15s %-3d'%(current.id,current.name,current.score))
      count +=1
      current=current.rlink
    print('********************************************')
    print('There is(are) %d record(s) found !!\n'%(count))

def main():
  option=0
  while True:
    print('\n*************Doubly linked list******************')
    print('                   <1>Insert                    ')
    print('                   <2>Delete                    ')
    print('                   <3>Modify                    ')
    print('                   <4>List                      ')
    print('                   <5>exit                      ')
    print('\n**********************************************')

    try:
      option=int(input('.  Choice:   '))
    except ValueError:
      print()
      print('No a correct number.')
      print('Try.again\n')

    if option==1:
      insert_f()
    elif option==2:
      delete_f()
    elif option==3:
      modify_f()
    elif option==4:
      display_f()
    elif option==5:
      break
    else:
      print('\n 不正確的選項')

main()
```
