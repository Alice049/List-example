# **鏈結串列的應用**
由於堆疊的加入和刪除都是同一端，因此可將它視為將節點加入和刪除都在串列的前端或尾端
佇列的加入和刪除為相異端，因此可以加入在前端，刪除在後端，或是相反。
其中
多項項相加可以用鏈結串列完成，因為多項式鏈結串列的資料結構表示成
COEF是變數的係數，EXP是變數的指數，LINK為指向下一個節點的指標。

```python
class polynomial:
  def __init__(self):
    self.coef=0
    self.exp=0
    self.next=None
#多項式1與多項式2與合併後的多項式ans_h

ptr=None
eq_h1=None
eq_h2=None
ans_h=None

def input_f():
  global ptr

  eq_h=None
  prev=None
  while True:
    ptr=polynomial()
    ptr.next=None

    try:
      ptr.coef=int(input('請輸入係數......'))
    except ValueError:
      print('Not a correct number.')
      print('Try again \n')

    if ptr.coef==0:
      return eq_h

    try:
      ptr.exp=int(input('請輸入指數......'))
    except ValueError:
      print('Not a correct number.')
      print('Try again\n')

    if eq_h==None:
      eq_h=ptr
    else:
      prev.next=ptr
    prev=ptr

def polynomialAddition():
  global ptr
  global ans_h

  this1=eq_h1
  this2=eq_h2
  prev=None

  while this1 !=None or this2 !=None:#當兩多項式皆相加完畢則結束
    ptr=polynomial()
    ptr.next=None
    #當第一個多項式指數大於第二個多項式
    if(this1 !=None and this2 ==None)or this1 !=None\
                    and this1.exp > this2.exp:
      ptr.coef=this1.coef
      ptr.exp=this1.exp
      this1=this1.next
    elif this1 ==None or this1.exp <this2.exp:
    #當第一個多項式指數小於第二個多項式
      ptr.coef=this2.coef
      ptr.exp=this2.exp
      this2=this2.next
    else:
      ptr.coef=this1.coef + this2.coef
      ptr.exp=this1.exp
      if this1 !=None:
          this1=this1.next
      if this2 !=None:
          this2=this2.next

    if ptr.coef !=0:
        if ans_h ==None:
            ans_h=ptr
        else:
            prev.next=ptr
        prev=ptr
    else:
        ptr=None

def showPolynomial(head,text):
  node=head
  print('%s'%(text),end='')
  while node !=None:
    print('%dx^%d'%(node.coef,node.exp),end='')
    if node.next !=None and node.next.coef >=0:
      print('+',end= '')
    node=node.next
  print()

def main():
  global eq_h1
  global eq_h2
  print('**********************************')
  print('----------多項式的格式為：ax^b----------')
  print('**********************************')
  print('\n<<第一個多項式（若係數為0,則結束)>>')
  eq_h1=input_f()
  print('\n<<第二個多項式（若係數為0,則結束)>>')
  eq_h2=input_f()
  polynomialAddition()
  print()
  showPolynomial(eq_h1,'第一個多項式為:')
  showPolynomial(eq_h2,'第二個多項式為:')
  print()
  showPolynomial(ans_h,'   相加結果為:  ')

main()
```
